```python
import random
import words
import hangman_stages


#function to choose a word randomly from the words_list
def choose_word():
    return random.choice(words.words_list)


#function to display the word while guessing...
def display_word(word, guessed_letters):
    displayed_word = ""
    for letter in word:
        if letter in guessed_letters:
            displayed_word += letter
        else:
            displayed_word += "_ "
    return displayed_word


#function to check if the guessed word matches the chosen random word
def is_game_over(word, guessed_letters):
    return all(letter in guessed_letters for letter in word)


#Main function to be executed...
def hangman():
    print("Welcome to Hangman!")
    
    # Choosing a word randomly from the word list
    chosen_word = choose_word()
    
    #letters for guessing the word
    guessed_letters = []
    
    #For Hangman no.of Chances = 6.( Head, Body, Two Hands, Two legs)
    chances = 6

    #While there are still chances to guess and game is not over...
    while chances > 0 and not is_game_over(chosen_word, guessed_letters):
        print("\nChances left:", chances)
        print(hangman_stages.stages[chances])
        print("Word:", display_word(chosen_word, guessed_letters))

        guess = input("Guess a letter: ").lower()
        
        #Checking condition for invalid input 
        if len(guess) != 1 or not guess.isalpha():
            print("Invalid input. Please enter a single letter.")
            continue
            
        # If user enters/inputs the guessed letter again...
        if guess in guessed_letters:
            print("You've already guessed that letter.")
            continue
            
        #appends the guessed letters to the list
        guessed_letters.append(guess)

        #User loses a chance for every incorrect guess
        if guess not in chosen_word:
            chances -= 1
            print("Incorrect guess!")
    
    #Chances are over and if user guesses the word correctly...
    if is_game_over(chosen_word, guessed_letters):
        print("\nCongratulations! You've guessed the word:", chosen_word)
        print(hangman_stages.stages[-1])
    #Chances are over and if user doesn't guess the word correctly
    else:
        print("\nGame over! The word was:", chosen_word)

        
        
if __name__ == "__main__":
    hangman()
```
